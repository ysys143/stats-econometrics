# I. 확률변수(Random Variable)와 확률분포



1. 확률변수 

확률변수 X는 표본공간 $\Omega$ 상에서 정의되는 real valued function이다. 확률변수는 어떤 확률적 실험의 결과로 나올 수 있는 모든 가능한 결과(표본공간)에 대해 실수공간(상태공간) 상의 값을 대응시킨다. 하나의 실험에 대해 여러 가지 확률변수가 정의될 수도 있는데, 가령 주사위를 던지는 실험에서 눈금의 값을 그대로 대응시키는 확률변수가 있을 수 있고, 짝수는 0, 홀수는 1을 대응시키는 확률변수가 있을 수 있다.



2. 확률밀도함수(PDF)와 누적분포함수(CDF)

확률변수는 확률분포를 동반하며, 확률분포는 확률(밀도)함수 또는 누적분포함수를 통해 표현될 수 있다.

확률함수(probability function)는 확률변수가 취하는 값이 이산적인 경우, $X$가 취하는 각 값에 대해 확률을 대응시켜준다.

확률밀도함수(PDF;probability density function)는 $X$가 취하는 값이 연속적인 경우, $X$가 취하는 값이 I라는 구간에 속할 확률을 해당 구간의 면적$P(X \in I) = \displaystyle \int_{I}f(x)dx$을 통해 제공한다.



누적분포함수(CDF;cumulative distribution function)는 $X$가 취하는 값이 특정값보다 낮을 확률을 제공한다. $F(x) \equiv P(X\leq x) = \displaystyle\int_{-\infty}^{x}f(t)dt$

3. 결합확률밀도함수(joint probability density function)



두개의 확률변수 X,Y가 이산적인 경우, 결합확률함수는 $f_xy(x_i, y_i) \equiv P(X=x_i, Y=y_i)$와 같이 주어지며, 연속적인 경우 결합확률밀도함수는 $f_xy((x, y) \in D) \equiv \displaystyle\iint_{D} f_xy(x,y) dxdy$와 같이 주어진다.



[4] 한계확률밀도함수(marginal probability density function)

다른 변수들을 고정했을 때 해당 변수의 한계확률밀도를 말한다.

$f_x(x_i) = \displaystyle\sum_{j} f_{xy}(x_i, y_i), \quad i=1,2,... \quad$ : X의 한계확률함수

$f_x(x_i) = \displaystyle\int_{-\infty}^{\infty} f_{xy}(x, y)dy \quad $ : X의 한계확률밀도함수



[5] 조건부확률밀도함수(conditional probability density function)



이산적인 경우 $Y=y_j$로 주어졌을 때, $X$의 조건부확률함수는 다음과 같이 주어진다.

$f_{x|y}(x_i | y_i) = P(X=x_i | Y=y_j) = \dfrac{P(X=x_i, Y=y_j)}{P(Y=y_j)} = \dfrac{f_{xy}(x_i, y_j)}{f_y(y_j)} = \dfrac{f_xy(x_i,y_j)}{\displaystyle\sum_{i} f_{xy}(x_i, y_j)}$



연속적인 경우 Y=y로 주어졌을 때, X의 조건부확률밀도함수는 다음과 같이 주어진다.

$f_{x|y}(x|y) = \dfrac{f_xy(x,y)}{f_{y}(y)} = \dfrac{f_xy (x,y)}{\displaystyle\int_{-\infty}^{\infty} f_xy(x,y) dx} $


# II. 수학적 기대값(Mathematical Expectation)

1. $E(X), V(X)$

$g(X)$가 확률변수 $X$의 함수라 하자. 

$E(g(X)) = \displaystyle\sum_{i} g(x_i)f(x_i) \quad $ (이산적인 경우)

$E(g(X)) = \displaystyle\int_{-\infty}^{\infty} g(x)f(x)dx \quad $ (연속적인 경우)

$g(X)=X$일 때, $E(g(X))=E(X)=\mu$를 확률변수 X의 평균이라 한다.

$g(X)=(X-E(X))^2$일 때, $E(g(X))=E[(X-E(X))^2]=V(X)=\sigma^2$을 확률변수 X의 분산이라 한다.

$\sigma = \sqrt{V(X)}$는 X의 표준편차라 한다.


2. $E(X,Y), Cov(X,Y)$

$g(X,Y)$가 확률변수 $X$,$Y$의 함수라 할 때,

$E(g(X,Y))=\displaystyle\sum_i \displaystyle\sum_j g(x_i, y_j)f(x_i, y_j) \quad $ (이산적인 경우)

$E(g(X,Y))=\displaystyle\int_{-\infty}^{\infty} \displaystyle\int_{-\infty}^{\infty} g(x, y)f(x, y) dx dy \quad$ (연속적인 경우)

$g(X,Y) = (X-\mu_X)(Y-\mu_Y)$일 때 $E[(X-\mu_X)(Y-\mu_Y)]=Cov(X,Y)$를 $X$와 $Y$의 공분산(Covariance)이라 한다.

3. 수학적 기대값의 성질

$V(X) = E(X^3. - [E(X)]^2$

$Cov(X,Y) = E(XY) - E(X)E(Y)$

$E(aX+bY+c)=aE(X)+bE(Y)+c$

$V(aY+bY+c)=a^2V(X)+b^2V(Y)+2ab Cov(X,Y) \quad$ (a,b,c는 상수)

$E(\displaystyle\sum_i a_i X_i) = \displaystyle\sum_i a_i E(X_i)$

$V(\displaystyle\sum_i a_i X_i) = \displaystyle\sum_i \displaystyle\sum_j a_i a_j Cov(X_i, X_j) \quad$ ($a_i$는 상수, 단, $Cov(X_i, X_i) = V(X_i))$

$\rho_{XY} = \dfrac{Cov(X,Y)}{\sqrt{V(X)}\sqrt{V(Y)}} \quad$ : 확률변수 X, Y 간의 상관계수(correlation coefficient)


# III. 확률적 독립(Stochstic Independence)

1. 확률적 독립의 정의

일반적으로 $n$개의 확률변수 $X_1, X_2, ... , X_n$이 서로 확률적으로 독립(mutually stochastically inpedendent)이라 함은, 각 변수의 한계확률밀도함수를 $f_1(x_1), f_2(x_2),...,f_n(x_n)$이라 할 때, $X_1, X_2, ... , X_n$의 결합확률밀도함수가 $f(x_1,x_2, ..., x_n)=f_1(x_1)f_2(x_2)...f_n(x_n)=\displaystyle\prod_i f_i(x_i)$와 같이 한계확률밀도함수의 곱으로 나타나는 경우로 정의된다.

2. 확률적 독립과 관련된 성질

확률변수 $X_1, X_2, ... , X_n$이 서로 확률적으로 독립일 경우,

$E\left\{ \displaystyle\prod_i^n u_i(X_i) \right\} = \displaystyle\prod_i^n E(u_i(X_i))$이 성립한다. 이를테면 $E(X_1 X_2) = E(X_1) E(X_2)$.

$Cov(X_i, X_j) = \rho_{X_i X_j} = 0$. 그러나 그 역은 성립하지 않는다.

$Var(\displaystyle\sum_i a_i X_i) = \displaystyle\sum_i a_i^2 V(X_i)$. 이를테면 $V(X_1 \pm X_2 ) = V(X_1) + V(X_2)$.

# IV. 이산형 확률분포들(discrete probability distributions)

1. 베르누이 실험(Bernouilli Experiment)：가능한 결과가 오직 두가지로 나타나는 실험. 
확률변수 $X$를 한번이 네르누이 실험에서 나오는 성공의 횟수라고 하고, 성공의 확률을 $p$라고 하면 $X$의 확률분포는 다음과 같다.

$P(X=1)=p, P(X=0)=1-p=q$. 달리 표현하면, $f(i) = p^i q^{1-i}, \quad i=0,1$.

$E(X)=p, \quad V(X)=pq$

2. 이항분포(binomial distribution): N번의 독립적 베르누이 실험시, 성공의 횟수

$f(i) = B(n,p) = \binom{N}{i}p^i q^{N-i}, \quad i=0,1,2,...,N$

$E(X)=Np, \quad V(X)=Npq$

3. 푸아송분포(Poisson distribution):　시간당 $\psi$의 비율로 물고기가 잡힐 때, $T$시간 동안 잡힐 물고기의 수.

$f(i) = P(\lambda) = \dfrac{\lambda^i e^{-\lambda}}{i!}$. 이때 $\lambda = T \times \psi$.

$E(X) =\lambda, \quad V(X)=\lambda$

$n$이 충분히 크고 $p$가 충분히 작은 값일 때, $B(n,p)$는 $P(np)$로 근사.

$\lambda_1$, $\lambda_2$의 모수를 갖는 두 독립적인 푸아송 분포의 함은 $\lambda_1+\lambda_2$를 모수로 하는 푸아송분포를 함.

4. 다항분포(multinomial distribution): 각 시행 당 3개 이상의 결과가 가능한 실험을 독립적으로 행할 때, A가 X번, B가 Y번 나올 확률.

$f(i,j) = \displaystyle\binom{n}{i,j,k} p_X^i p_Y^j p_Z^k, \quad i=0,1,...,n, j=0,1,...,n-i$. 단, $k=n-i-j, p_z = 1-p_X-p_Y$.

$Cov(X,Y) = -np_X p_Y$

X와 Y의 한계확률분포는 각각 이항분포 $B(n, P_X), B(n, P_Y)$가 된다.


# V. 연속형 확률분포들(continuous probability distributions)

## [기본 분포들]

1. 균등분포(uniform distribution)

$f(x) = U(a,b) =\dfrac{1}{b-a}, a<x<b \quad$ 그 밖의 구간에서는 0.

$F(X) =
\begin{cases} 
0 & \text{if } x \leq a \\
\dfrac{x - a}{b - a} & \text{if } a < x < b \\
1 & \text{if } x \geq b
\end{cases}$

$E(X) = \dfrac{a+b}{2}$

$V(X) = \dfrac{(b-a)^2}{12}$

2. 기하분포(exponential distribution)

물고기 한마리 낚는데 걸리는 시간이 $\beta$일 때(또는 시간당 $\psi=1/\beta$의 비율로 물고기가 낚임), 첫 물고기를 낚을 때까지 걸리는 시간.

$f(x) = E(x) = \dfrac{1}{\beta}\exp(-\dfrac{x}{\beta}), x>0$

$F(X) =
\begin{cases} 
0 & \text{if } x \leq 0 \\
1-\exp(-\dfrac{x}{\beta}) & \text{if } x > 0
\end{cases}$

$E(X) = \beta$

$V(X) = \beta^2$

3. 정규분포(normal distribution)

$f(x) = N(\mu, \sigma^3. = \dfrac{1}{\sqrt{2\pi}\sigma}\exp[-\frac{1}{2}(\frac{x-\mu}{\sigma})^2] \quad -\infty < x < \infty$

$E(X) = \mu$

$V(X) = \sigma^2$

$X_1 \sim N(\mu_1, \sigma_1^3., X_2 \sim N(\mu_2, \sigma_2^3.$이면

$aX_1 + bX_2 \sim N(a\mu_1 + b\mu_2, a^2 \sigma_1^2 + b^2 \sigma_2^2 + 2ab \sigma_{12}) $

## [변환을 통해 도출되는 분포들]

1. $\chi^2$ 분포 : 정규분포를 따르는 변수들의 제곱합의 분포

$X = \displaystyle\sum_{i=1}^{n} Z_i^2$ ($Z_1, ..., Z_n$은 서로 독립인 표준정규분포)

$f(x) = \chi_n^2(x) = \dfrac{x^{\frac{n}{2}}e^{-\frac{x}{2}}}{2^{\alpha}{\Gamma(\frac{n}{2})}} \quad x>0$

$E(X) = n$

$V(X) = 2n$

2. t-분포 : 정규분포를 카이제곱 분포로 나누고 $\sqrt{n}$을 곱한 분포

$X = \dfrac{Z}{\sqrt{V/n}}$

이때 $Z \sim N(0,1), V \sim \chi_n^2$이며 $Z, V$는 서로 독립.

$f(x) = t_n(x) = \dfrac{  \Gamma(\frac{n+1}{2}) } { \Gamma(\frac{n}{2}) \sqrt{n\pi} } \dfrac{1}{ {(1 + \frac{x}{n})}^{\frac{n+1}{2}} } \quad  -\infty < x < \infty$

$E(X)=0 \quad $ ($n\geq2$인 경우에만 존재)

$V(X)=\dfrac{n}{n-2} \quad $  ($n\geq3$인 경우에만 존재)

자유도가 1인 t분포는 표준코시분포가 됨.

3. F-분포 : 서로 독립인 두 카이제곱분포가 있을 때, 두 값의 비율의 분포

$X = \dfrac{V_1/n}{V_2/m}$

$F_{n,m} = 1/F_{m,n}$

$t_n^2 = F_{1,n}$

이때 $V_1 \sim \chi_n^2, V_2 \sim \chi_m^2$이며 $V_1, V_2$는 서로 독립.

$f(x) = F_{n,m} = \dfrac{  \Gamma(\frac{n+m}{2}) } { \Gamma(\frac{n}{2}) \Gamma(\frac{m}{2})  } \left(\dfrac{n}{m}\right)^{\frac{n}{2}} \dfrac{ x^{\frac{n}{2}-1}}{ {(1 + \frac{n}{m} x)}^{\frac{n+m}{2}} } \quad x>0$

$E(X) = \dfrac{m}{m-2} \quad$  ($n\geq3$인 경우에만 존재)

$V(X) = \dfrac {2m^2(n+m-3.} { n (m-3.^2(m-4. } \quad $  ($n\geq5$인 경우에만 존재)

$F_{n,m} \sim \dfrac{1}{F_{m,n}}$

$t_n^2 = \dfrac{Z^2}{V_n/n} = F_{1,n}$
