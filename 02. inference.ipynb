통계적 추론

I. 확률표본(random sample)

어떤 특정한 분포로부터 '크기가 n인 확률표본'은 n개의 동일한 특정 분포를 갖는 서로 확률적으로 독립인 확률변수 $S = \{X_1, X_2, ..., X_n\}$. 확률표본 또한 확률변수라는 것.

통계량 : 미지의 모수에 의존하지 않고 표본 $S$를 어떤 실수값에 대응시키는 함수로 정의할 수 있다. $T(S): \mathbb{R}^n \to \mathbb{R}$.

통계치 : 확률표본의 값들이 관찰될 경우, 그 값들에 대응하는 통계량의 값을 계산할 수 있으며 이를 통계치(a statstic)이라 부른다.

1. 표본평균(sample mean)

$\bar{X_n} = \dfrac{\sum_1^n X_i}{n}$

$E(X_i) = \mu, \quad Var(X_i) = \sigma^2 \quad \Rightarrow  \quad  E(\bar{X})=\mu , \quad Var(\bar{X})$


2. 표본분산(sample variance)

$S^2 = \dfrac{\sum_1^n (X_i - \bar{X})^2}{n-1} \quad$ 또는 $\quad S^{*2} = \dfrac{\sum_1^n (X_i - \bar{X})^2}{n}$ 

$E(S^2)=\sigma^2 \quad, \quad E(S^{*2})= \dfrac{n-1}{n} \sigma^2$

3. $N(\mu, \sigma^2)$로부터의 확률표본

표본평균 $\bar{X} \sim N \left( \mu, \sigma^2/n \right) $

표본분산 $\dfrac{n-1}{\sigma^2}S^2 = \dfrac{\sum_1^n(X-\bar{X})^2}{\sigma^2} \sim \chi_{n-1}^2 $ : 자유도가 n-1인 카이제곱분포

$\bar{X} \perp S^2$ : 표본평균과 표본분산은 확률적으로 독립이다 => t분포를 따르는 통계량 도출 가능

II. 극한분포이론(Limiting Distribution Theory)

1. 대수의 (약)법칙 = 확률적 수렴(convergence in probability)

i.i.d(independent and identically distributed)인 $X_1, X_2, ..., X_n$가 있고, 모집단의 평균과 분산이 존재한다고 하자. $E(X_i)=\mu<\infty, V(X_i)=\sigma^2<\infty$.

임의의 양의 실수 $\epsilon$에 대해, $P(|\bar{X_n}-\mu|>\epsilon) \to 0$ as $n \to \infty$. 다시 말해 $\bar{X_n}$는 $\mu$로 확률적으로 수렴한다. $\bar{X_n} \overset{p}{\to} \mu$. $\quad p \displaystyle \lim_{n \to \infty} \bar{X_n} = \mu$.

서로 독립적이지 않고 동일한 분포를 하지도 않는 $X_1, X_2, ..., X_n$에 대해서도, 모집단의 평균과 분산이 존재하기만 한다고 하자. $E(\sum_1^n X_i)=\mu<\infty, V_n = V(\sum_1^n X_i)=\sigma^2<\infty$. 

여기에 $V_n/n^2 \to 0 \,$ as $\, n \to \infty$이라는 조건만 추가된다면, $p \displaystyle \lim_{n \to \infty} \bar{X_n} = \dfrac{m_n}{n}$.

약법칙은 점근적으로 가까워진다는 것이고, 강법칙은 반드시 수렴한다는 것이다.

2. 중심극한정리(CLT; Central Limit Theorem)

확률변수 $X_n$의 분포함수의 값이 모든 연속적인 점에서 $n \to \infty$에 따라 확률변수 Y의 분포함수의 값으로 수렴할 때, 확률변수 $X_n$이 확률변수 $Y$로 분포수렴한다고 한다. $X_n \overset{D}\to Y$.

i.i.d(independent and identically distributed)인 $X_1, X_2, ..., X_n$가 있고, 모집단의 평균과 분산이 존재한다고 하자. $E(X_i)=\mu<\infty, V(X_i)=\sigma^2<\infty$.

이때 다음이 성립하여 중심극한정리라 부른다. 대수의 법칙이 표본의 수가 많아질 수록 표본평균이라는 통계량이 모평균이라는 한 점으로 수렴함을 말하는 반면, 중심극한정리는 그렇게 표본평균의 '분산'이 점차 좁아지는 과정에서도 $\sqrt{n}$을 곱해 보정해주면 그 분포가 정규분포 $N(\mu, \sigma^2)$로 근사할 수 있게 됨을 말한다.

$\dfrac{\bar{X_n}-\mu}{\sigma/\sqrt{n}} \overset{D}\to N(0,1) \quad$ as $\quad n \to \infty$

서로 독립적이지 않고 동일한 분포를 하지도 않는 $X_1, X_2, ..., X_n$에 대해서도,  $\displaystyle \lim_{n \to \infty} nV(\bar{X_n}) = \sigma^2 < \infty$ 모분산의 기대값이 수렴한다고 가정하면 다음이 성립한다.

$\dfrac{\bar{X_n}-E(\bar{X_n})}{\sqrt{V(\bar{X_n})}} \overset{D}\to N(0,1) \quad$ as $\quad n \to \infty$


III. 점추정(Point Estimation)

1. 추정량(Estimator)

모수 $\theta$의 추정을 위해 적절히 고안된 통계량 $\hat{\theta}_n=\hat{\theta}(X_1,X_2,...,X_n)$을 추정량, 관측된 표본 값에 대응되는 해당 통계량의 값을 추정치(Estimate)라 함. 표본집합 S의 함수라는 것.

추정량에 요구되는 성질로 불편성, 일치성, 유효성이 있다.

1.1 불편성(unbiasedness) 또는 점근적 불편성(asymptotic unbiasedness)

추정량의 기대값이 실제 모수와 동일한 성질. 추정량이 모수를 과대 또는 과소 추정하지 않는다는 것을 의미하며, 추정량이 "장기적으로" 모집단의 진짜 값을 추정한다는 성질을 가지고 있다.

$E(\hat{\theta})=\theta \,$(불편추정량) $\quad$ 또는 $\quad \displaystyle \lim_{n \to \infty} E(\hat{\theta})=\theta \,$(점근적 불편추정량)

1.2 일치성(consistency)

표본 크기가 증가할수록 추정량이 실제 모수에 수렴하는 성질. 표본의 크기를 계속해서 증가시키면 추정량이 모수에 수렴한다는 것을 의미한다. 

일치성을 만족한다면, $S^*$와 같이 불편성은 만족하지 않을 수도 있지만 점근적 불편성은 반드시 만족하게 됨.

$\hat{\theta}_n \overset{P}\to \theta \,$(일치추정량) $\quad \Rightarrow \quad$ (점근적) 불편추정량이고, $\, \displaystyle \lim_{n \to \infty} V(\hat{\theta})=0$

1.3 유효성(efficiency)

불편 추정량들 중에서 가장 작은 분산을 가지는 추정량. 즉, 같은 모수를 추정하는 여러 불편 추정량 중에서 분산이 가장 작은 추정량을 유효한 추정량이라고 한다.

추정량 $\hat{\theta}_1$과 $\hat{\theta}_2$가 불편 추정량일 때, $Var(\hat{\theta}_1)<Var(\hat{\theta}_2)$이라면 , $\hat{\theta}_1$이 $\hat{\theta}_2$보다 더 유효한 추정량이다.

크래머-라오 하한(CR Lower Bound; CRLB)은 피셔 정보 $I(\theta)$의 역수로서, 불편추정량 $Var(\hat{\theta})$의 이론적 하한을 제공한다. $\hat{\theta}_1$의 유효성은 $\left ( CRLB/V(\hat{\theta}) \right ) \times 100\%$로 나타낸다.

100% 유효성을 갖는 불편추정량을 유효추정량이라 하며, 점근적 불편추정량의 분산이 점근적으로 CRLB에 도달할 경우 점근적 유효추정량이라 한다.

불편추정량과 점근적 불편추정량의 성능(Performance)은 보통 평균제곱오차(Mean Squared Error; MSE)로 비교한다.

$MSE(\hat{\theta}) = V(\hat{\theta}) + [E(\hat{\theta})-\theta]^2$

유효추정량은 최소분산불평추정량(Minimum Variance Unbiased Estimator; MVUE)이다.

2. 적률방법(Method of moments)

적률방법 추정은 모집단 적률(보통 모수의 함수)과 표본 적률을 같게 만드는 방정식을 세워서, 모수를 추정하는 방식이다. 적률방법추정량은 점근적 불편성과 일치성을 가진다.

$\mu_k = E(X^k) = \displaystyle\int_{-\infty}^{\infty} x^kf(x)dx$ : k번째 모집단 적률

$\hat{\mu}_k = \dfrac{1}{n}\sum_1^n(X_i)^k $ : k번째 표본 적률 


적률방법 추정량(MMEs)의 계산 방법 적률방법은 다음과 같다.  모집단의 적률을 모수의 함수로 나타내보자. 

모집단이 정규분포 $N(\mu,\sigma^2)$를 따른다면, 1차 모집단 적률(평균)은 $\mu_1 = E(X) = \mu$고, 

2차 모집단 적률(분산)은 $\mu_2 = E(X^2) = [E(X)]^2  + E(X^2) - [E(X)]^2= \mu^2 + \sigma^2$이다. 

이제 실제 데이터로부터 표본 적률 $\hat{\mu}_k$를 계산하고 모집단 적률과 표본 적률을 같게 만드는 방정식을 세운다. 

먼저, 1차 표본 적률과 2차 표본 적률은 다음과 같이 주어진다.

$\hat{\mu}_1 = \dfrac{1}{n}\sum_{i=1}^{n}X_i$

$\hat{\mu}_２ = \dfrac{1}{n}\sum_{i=1}^{n}X_i^2$

이제 1차 표본 적률과 1차 모집단 적률을 같게 둬보자. 

$\mu_1 = \hat{\mu}_1 = \hat{\mu}$

$\dfrac{1}{n}\sum_{i=1}^{n}X_i = \hat{\mu}$

2차 표본 적률과 2차모집단 적률을 같게 두고, 다른 모수도 추정할 수 있다.

$\mu_２= \hat{\mu}_２ = \hat{\mu}^2+\hat{\sigma}^2$

$\hat{\mu}_２ =  \dfrac{1}{n}\sum_{i=1}^{n}X_i^2 = \hat{\mu}^2+\hat{\sigma}^2 = [\dfrac{1}{n}\sum_{i=1}^{n}X_i^2]^2 + \hat{\sigma}^2$

$\hat{\sigma}^2 =  \dfrac{1}{n}\sum_{i=1}^{n}X_i^2 - [\dfrac{1}{n}\sum_{i=1}^{n}X_i^2]^2$



3. 최우추정법(Maximum Likelihood Method)

3.1 우도함수(Likelihood function) : 모수의 함수로 나타낸 표본의 결합확률분포.

$L(\theta_1, \theta_2,...,\theta_k)= \displaystyle\prod_1^n f(x_i|\theta_1, \theta_2,...,\theta_k)$

최우추정법은 우도함수를 극대화시키는 모수의 값을 해당 모수의 추정치로 삼는 방법이며, 이를 확률표본 $X_1, X_2, ... , X_n$의 함수로 볼 때, 최우추정량(Maximum Liklihood Estimator; MLE)라 함. 

우도함수의 극대화 해를 바로 찾는 것보다는 우도함수에 로그를 씌운 $\ln L$을 극대화하는 해를 찾는 것이 보통이다.

최우추정법을 통해 구한 $\theta$의 추정량 $\hat{\theta}_n = \hat{\theta}(X_1, X_2, ..., X_n)$은 점근적 불편성, 일치성, 그리고 점근적 유효성을 갖는다. 

$nV(\hat{\theta}_{n\sigma}) \to \sigma(\theta)$ as $n \to \infty$, 단 $\sigma(\theta)/n$은 CRLB. &nbsp; (점근적 유효성)

충분히 큰 n에 대해 $\hat{\theta}_n \sim N \left( \theta, \sigma^2(\theta)/n \right)$으로 근사할 수 있다.

$\sigma(\theta)$에 대한 추정량을 $\sigma(\hat{\theta}_n)$라 하면, $\hat{\theta}_n \sim N \left( \theta, \sigma^2(\hat{\theta}_n)/n \right)$으로 근사하여 사용할 수 있다.

IV. 신뢰구간(Confidence Interval; CI)　또는 구간추정(Interval Estimation)

1. 모집단이 $N(\mu, \sigma^2)$일 때, $\mu$에 대한 $100 \times (1-\alpha) \%$  CI

1) $\sigma^2$이 알려진 경우  

CI : $\bar{X} \pm Z_{\alpha/2} \dfrac{\sigma}{\sqrt{n}} \quad$단,$\,\, P(Z>z_{\alpha}) = \alpha \,,\,Z \sim N(0,1)$

2) $\sigma^2$를 모르는 경우  

CI : $\bar{X} \pm t_{\alpha/2, n-1} \dfrac{S}{\sqrt{n}} \quad$단,$\,\, P(T_{n-1}>t_{\alpha,n-1}) = \alpha \,,\,T_{n-1}$는 자유도가 $n-1$인 $t$ 분포

2. 모집단이 $N(\mu, \sigma^2)$일 때, $\sigma^2$에 대한 $100 \times (1-\alpha) \%$ CI

CI : $\left( \dfrac{(n-1)S^2}{\chi^2_{1-\alpha/2,n-1}}, \dfrac{(n-1)S^2}{\chi^2_{\alpha/2,n-1}} \right) \quad$단,$\,\, P(\chi^2_{n-1} > \chi^2_{\alpha, n-1} ) = \alpha \,,\,\chi^2_{n-1}$은 자유도 $n-1$인 $\chi^2$ 분포.



V. 가설검정(Hypothesis Testing)

1. 가설검정의 구성요소

1.1 귀무가설(null hypothesis)과 대립가설(alternative hypothesis)

귀무가설($H_0$)은 검정결과 기각해야 하는 증거가 나타나기 전까지 사실로 받아들여야 하는 가설

대립가설($H_1$)은 검정결과 귀무가설을 기각하는 경우 사실로 받아들여야 하는 가설(증거를 통해 증명하고자 하는 새로운 사실이나 관계)

1.2 검정통계량(test statistic)

귀무가설 하에서 (적어도 근사적으로라도) 확률분포가 알려진 통계량 $T = T(X_1, X_2, ..., X_n)$

검정통계량의 대립가설 하의 분포는 알려져있을 필요는 없으나 귀무가설 하의 분포와는 구별되는 분포를 해야 함.

1.3 기각역(rejection region)

검정통계량의 값이 이 구간에 포함되면 귀무가설을 기각함

2. 1종 오류(Type I error)와 2종 오류(Type II error)

2.1 1종 오류 : 모집단에서 귀무가설이 옮음에도 표본으로부터 귀무가설을 기각하는 경우

제1형 오류를 범할 확률을 검정의 크기(size)　또는 유의수준(significance level)이라고 함($\alpha$). 보통 1%, 5%, 10% 등으로 주어진다.

$\alpha = P(\,T(X_1,X_2,...,X_n)\in R \,|\, H_0 = True)$

2.2 2종 오류 : 모집단에서 귀무가설이 옮지 않음에도 표본으로부터 귀무가설을 기각하지 못하는 경우

귀무가설이 옳지 않을 때 귀무가설을 기각할 확률을 검정력(power of test)라 하며, 2종 오류의 확률을 $\beta$라 하면, 검정력은 $1-\beta$.

$1-\beta=P(\,T(X_1, X_2, ..., X_n) \in R \,|\, H_0 = True) \,$: 검정력 함수


3. CI에 근거한 검정

각 경우에 있어서 CI를 도출하는 과정을 뒤집어 적절한 검정통계량을 고안해낼 수 있음

$N(\mu,\sigma^2)$일 때 $\mu$에 대한 $100 \times (1-\alpha) \%$ CI는 $\bar{X} \pm t_{\alpha/2, n-1} \dfrac{S}{\sqrt{n}}$

$P \left( \left| \dfrac{\bar{X}-\mu}{S/\sqrt{n}} \right| > t_{\alpha/2,n-1}  \right) = 1-\alpha \quad \Rightarrow \quad \dfrac{\bar{X}-\mu}{S/\sqrt{n}} \sim t_{n-1}$을 이용하여 $\mu$와 관련된 가설검정의 검정통계량으로 이용 

3.1 양측검정

$H_0:\mu=\mu_0 \,,\, H_1:\mu \neq \mu_0$ : $\left| \dfrac{\bar{X}-\mu}{S/\sqrt{n}} \right| > t_{\alpha/2,n-1}$이면 유의수준 $\alpha$에서 귀무가설을 기각($t_{\alpha}$가 아니라 $t_{\alpha/2}$임에 유의).

3.2 단측검정(우측검정)

$H_0:\mu=\mu_0 \,,\, H_1:\mu > \mu_0$ (또는 $H_0:\mu \leq \mu_0 \,,\, H_1:\mu > \mu_0$  ) $ \dfrac{\bar{X}-\mu}{S/\sqrt{n}}  > t_{\alpha,n-1}$이면 유의수준 $\alpha$에서 귀무가설을 기각.

3.3 단측검정(좌측검정)

$H_0:\mu=\mu_0 \,,\, H_1:\mu < \mu_0$ (또는 $H_0:\mu \geq \mu_0 \,,\, H_1:\mu < \mu_0$  ) $ \dfrac{\bar{X}-\mu}{S/\sqrt{n}}  < t_{\alpha,n-1}$이면 유의수준 $\alpha$에서 귀무가설을 기각.
