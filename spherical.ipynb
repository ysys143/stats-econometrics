```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randn(2, 10000)

# 2  x[0] x[1] 
plt.scatter(x[0], x[1], color='blue', s=2)  #   
plt.xlim(-5, 5)  # x  
plt.ylim(-5, 5)  # y  
plt.axhline(0, color='black',linewidth=0.5, ls='--')  # x
plt.axvline(0, color='black',linewidth=0.5, ls='--')  # y
plt.title('2D Visualization of x[0] and x[1]')  #  
plt.grid()  #  
plt.show()  #  

```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # 3D    

# x 3  
x = np.random.randn(3, 10000)

# 3D  x[0], x[1], x[2] 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')  # 3D  
ax.scatter(x[0], x[1], x[2], color='blue', s=1)  #     50 
ax.set_xlim(-5, 5)  # x  
ax.set_ylim(-5, 5)  # y  
ax.set_zlim(-5, 5)  # z  
ax.set_title('3D Visualization of x[0], x[1], and x[2]')  #  
plt.show()  #  
```
