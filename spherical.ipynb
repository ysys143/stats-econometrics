```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randn(2, 10000)

# 2차원 공간에 x[0]과 x[1] 시각화
plt.scatter(x[0], x[1], color='blue', s=2)  # 점을 파란색으로 표시
plt.xlim(-5, 5)  # x축 범위 설정
plt.ylim(-5, 5)  # y축 범위 설정
plt.axhline(0, color='black',linewidth=0.5, ls='--')  # x축
plt.axvline(0, color='black',linewidth=0.5, ls='--')  # y축
plt.title('2D Visualization of x[0] and x[1]')  # 제목 설정
plt.grid()  # 그리드 추가
plt.show()  # 그래프 표시

```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # 3D 시각화를 위한 모듈 추가

# x를 3차원 배열로 생성
x = np.random.randn(3, 10000)

# 3D 공간에 x[0], x[1], x[2] 시각화
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')  # 3D 축 추가
ax.scatter(x[0], x[1], x[2], color='blue', s=1)  # 점을 파란색으로 표시하고 크기를 50으로 설정
ax.set_xlim(-5, 5)  # x축 범위 설정
ax.set_ylim(-5, 5)  # y축 범위 설정
ax.set_zlim(-5, 5)  # z축 범위 설정
ax.set_title('3D Visualization of x[0], x[1], and x[2]')  # 제목 설정
plt.show()  # 그래프 표시
```
